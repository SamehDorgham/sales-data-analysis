# **1- What was the total revenue generated by the company over the course of the year?**

# **2- Which product had the highest revenue? How much revenue did it generate?**

```python
MaxProductRevenueId = data["revenue"].idxmax()
MaxProductRevenueId
```

```python
MaxProductRevenue = data["product"][MaxProductRevenueId]
MaxProductRevenue
```

```python
DataLength = data["revenue"].count()
DataLength = int(DataLength)
DataIndex = []
for i in range(0,DataLength+1):
  DataIndex.append(i)
data["index"] = DataIndex
data.set_index("index", inplace=True)
data
```

```python
MaxProductRevenueValue = data["revenue"][MaxProductRevenueId]
MaxProductRevenueValue
```




---




# **3- What was the average price of a product sold by the company?**

```python
AveragePrice = data["price"].mean()
AveragePrice.round(2)
```




---




# **4- What was the total quantity of products sold by the company?**

```python
TotalQuantitySold = data["quantity"].sum()
TotalQuantitySold
```




---




# **5- Which category had the highest revenue? How much revenue did it generate?**

```python
CategoryWithHighestRevenue = data.groupby("category")["revenue"].sum()
CategoryWithHighestRevenue
```




---




# **6-What was the average revenue per sale?**


```python
AverageRevenuePerSale = data["revenue"].mean()
AverageRevenuePerSale.round(2)
```




---




# **7- What was the total revenue generated in each quarter of the year? (i.e. Q1, Q2, Q3, Q4)**

```python
data.describe()
```

```python
TotalQuarterRevenue = data["revenue"].groupby(data["date"].dt.quarter).sum()
TotalQuarterRevenue
```




---

# **Data Visualization Section**

**1- Revenue Over Time**

```python
X = data["date"]
Y = data["revenue"]
plt.plot(X,Y)
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.title("Revenue Over Time")
plt.legend(["Revenue"])
plt.show()
```



---



**2- Category With Highest Revenue**

```python
Category = CategoryWithHighestRevenue.index
Revenue = CategoryWithHighestRevenue.values
plt.bar(Category,Revenue)
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.title("Category With Highest Revenue")
plt.legend(["Revenue"])
plt.show()
```




---



**3- Quantity Sold Over Time**

```python
QuantitySold = data["quantity"]
Date = data["date"]
plt.bar(Date,QuantitySold)
plt.xlabel("Date")
plt.ylabel("Quantity Sold")
plt.title("Quantity Sold Over Time")
plt.legend(["Total Quantity Sold"])
plt.show()
```




---



**4- Total Quarter Revenue**

```python
X = TotalQuarterRevenue
Y = TotalQuarterRevenue.index
plt.bar(Y,X)
plt.xlabel("Quarter")
plt.ylabel("Revenue")
plt.title("Total Quarter Revenue")
plt.legend(["Revenue"])
plt.show()
```
